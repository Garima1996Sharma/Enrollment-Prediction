## 1.1 Load the python library
All the used libraries are mentioned below:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sb
```

## 1.2 Read the data

```python
def load_course_data(filename):
    return pd.read_csv(filename)

# data is loaded here.
df = load_course_data("course_data_uncleaned.csv")
df

df.head()
# provide the summary of the data types
df.info()
```

## 1.3 Check null values

```python
# for null values, return value in boolean with true for non-null and false for null
df.isna()
# will return the sum of null values per column
df.isna().sum()
```

## 1.4 Dropping and Filling data

```python
# Checking null values per column:
df.isna().sum()
```

```python
# dropping rows where all data are missing:
df.dropna(axis = 0, how = 'all', inplace = True)
print(df)
```

```python
# dropping the column price as it has more null values than price2.
# Checking null values per column:
df.isna().sum()

# column having more null values is price
df.price2.fillna(df.price,inplace=True)
#print(df)
df.isna().sum()

# drop column price
df.drop('price', inplace=True, axis =1)
df
```

```python
# dropping rows where all data are missing:
df.dropna(axis = 0, how = 'all', inplace = True)
df

df.head()
```

## 1.5 Changing Columns Label

```python
#using lambda function to change column name
df.rename(columns=lambda x: x.replace(" ","_"),inplace=True)
df
```

## 1.6 Remove special characters from the data and format some column data


```python
# removing % sign from discount table and CA$ from price2 column

df['discount'] = df['discount'].str.replace('\%','',regex=True)
df['price2'] = df['price2'].str.replace('CA', '',regex=True)
df['price2']=df['price2'].str.replace('$', '',regex=True)
df
```

```python
# transforming the duration into hours

print(df['duration'].dtypes)

# replacing h from duration
df['duration'] = df['duration'].str.replace('h','',regex=True)
# replacing m from duration
df['duration'] = df['duration'].str.replace('m','',regex=True)
print(df['duration'])
df['duration'].str.split(' ',expand=True)

hrs = df['duration'].str.split(' ',expand=True)[0].astype(float)
mins = df['duration'].str.split(' ',expand=True)[1].astype(float)

# function to calculate the hours
def hoursCalculate(hrs,mins):
     hours= (hrs+(mins/60))
     return hours.round(2)


df['duration'] = hoursCalculate(hrs, mins)
df
```

```python
# transforming the last_updated column into a number that represents the value

month= df['last_updated'].str.split('/', expand=True)[0].astype(float).round(2)
b= df['last_updated'].str.split('/', expand=True)[1]
year= b.str[-2:].astype(float).round(2)

# function is created to calculate the hours
def lastCalculate(a,c):
     last_updated = (c + (a/12))
     return last_updated.round(2)

df['last_updated'] = lastCalculate(month, year)
df
```

```python
df.dtypes

# changing the data type of columns discount, price2 and last_updated into float64
df['price2'] = df['price2'].astype(np.float64)
df['discount'] = df['discount'].astype(np.float64)
df.dtypes
```

## 1.7 Reduce the number of unique values in the course_category

```python
# checking the unique values
df['course_category'].unique()
# counting the uniques values
df['course_category'].value_counts()
```

```python
# combining any category containing the office, teaching and personal development to the business & education category. 
df.loc[
    df['course_category']
    .str.contains
    ("business|finance accounting|marketing|office productivity|teaching academics|personal development"), 'course_category'] = 'business & education'
```

```python
# combining the category music, lifestyle, health, photo, and other into personal hobby category. 
df.loc[
    df['course_category']
    .str.contains
    ("music|lifestyle|marketing|health fitness|photography|other"), 'course_category'] = 'hobby'

```

```python
#using lambda function to change column course_category whitespace
df['course_category'] = df['course_category'].replace(' ', "_", regex=True)

df
```

```python
# checking the count of categories
df['course_category'].value_counts()
```

## 1.8 Drop the unneeded columns and reset the index

```python
# dropping the unwanted columns
df.drop(['course_name', 'course_sub_category'], axis=1, inplace = True)
# reseting the index
df.reset_index(inplace=True, drop=True)
df
```

## 1.9 Saving the Cleaned Dataset

```python
# cleaned dataset is saved in csv file
df.to_csv("course_data_cleaned.csv")
```
