## 1.1 Load the python library
All the used libraries are mentioned below:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sb
```

## 1.2 Read the cleaned data

```python
def load_course_data(filename):
    return pd.read_csv(filename)

# data is loaded here.
df = load_course_data("course_data_cleaned.csv")
df

# drop column unnamed
df.drop('Unnamed: 0', inplace=True, axis =1)
df
print("\n")

# checking for null rows/columns
print(df.isna().sum())

# dropping rows where any data are missing:
df.dropna(axis = 0, how = 'any', inplace = True)

print("\n")
print(df.head())

print("\n")
# provide the summary of the data types
df.info()
print("\n")

# reseting the index
df.reset_index(inplace=True, drop=True)
df
```

## 1.6 Multi/Bivarite Analysis

```python
# relationship between enrollment and number_rating
mp.scatter(df.number_ratings, df.enrollment, color="orange")
mp.title("Enrollment vs number_rating of courses")
mp.xlabel("Number_rating")
mp.ylabel("Enrollment")
```

We can see that as the number of ratings increases, enrollment in courses also increasing. There are more enrollments when ratings lie between 0 to 4000 approximately.

```python
# relationship between enrollment and reviews
mp.scatter(df.inst_review, df.enrollment, color="orange")
mp.title("Enrollment vs review of courses")
mp.xlabel("Inst_reviewing")
mp.ylabel("Enrollment")

```

We can see that more enrollments when there are approximately reviews from 0 to 25000. After that, reviews are steady.

```python
# relationship between enrollment and average_rating
mp.scatter(df.avg_rating, df.enrollment, color="orange");
mp.title("Enrollment vs average rating of courses", color="red");
mp.xlabel("Avg_rating", color="green");
mp.ylabel("Enrollment", color="green");
```

We can see that enrollments are higher when average rating is within the range of 4.2 to 4.7 approximately.

## 1.9 Feature Engineering

```python
# transforming course category into dummy values
dummy_features = pd.get_dummies(df.course_category, drop_first = True, prefix = 'cat')
dummy_features

# dropping the unwanted column
df.drop(['course_category'], axis=1, inplace = True)
# reseting the index
df.reset_index(inplace=True, drop=True)
df

df = pd.concat([df,dummy_features], axis = 1)
df
```
